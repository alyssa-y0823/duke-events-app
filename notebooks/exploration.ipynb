# Ranking Logic Exploration

This notebook demonstrates the semantic search capability using Sentence-Transformers.

```python
import sys
import os
import numpy as np

# Add root path to access models
sys.path.append('../')

from models.embeddings import Embedder
```

```python
embedder = Embedder()
print("Model loaded successfully")
```

```python
query = "artificial intelligence"
events = [
    "Introduction to Machine Learning",
    "Painting Workshop",
    "Advanced Robotics"
]

q_emb = embedder.embed_text(query)
event_embs = embedder.embed_texts(events)

for event, e_emb in zip(events, event_embs):
    score = np.dot(q_emb, e_emb)
    print(f"Validation: '{query}' vs '{event}' = {score:.4f}")
```
